```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load dataset
file_path = "C:/Users/Shuhaib/Downloads/shuaib/DATA SET/public-transport-accessibility-smart-cities.csv"
data = pd.read_csv(file_path)

# Preprocessing: Selecting numerical columns for dimensionality reduction
numerical_columns = ["No. of buses of that type", "No. of bus terminals", "No. of bus stands", "No. of bus stops"]
data_numerical = data[numerical_columns].apply(pd.to_numeric, errors='coerce').dropna()

# Standardize the numerical data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numerical)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reducing to 2 dimensions for visualization
reduced_data = pca.fit_transform(data_scaled)

# Create a DataFrame for the reduced dimensions
reduced_df = pd.DataFrame(reduced_data, columns=["PCA1", "PCA2"])

# Visualize the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(reduced_df["PCA1"], reduced_df["PCA2"], c='blue', alpha=0.7, label="Cities")
plt.title("PCA Dimensionality Reduction", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=14)
plt.ylabel("Principal Component 2", fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
